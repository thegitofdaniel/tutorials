## Load data

```python
# equity: Tesla stock

# method 1
tsla = Stock('tsla')
df = pd.DataFrame(tsla.get_historical_prices())
df.head()
```

```python
# method 2
start = datetime(2018, 1, 1)
end = datetime(2022, 5, 9)
df = get_historical_data("TSLA", start, end, close_only=True, output_format='pandas')

print(df.shape)
print(df.head())
```

# Graphs
## Stock Price

```python
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(df.index, df.close)
plt.show()
```

## Simple Moving Average (MA)

```python
# series
rolling_mean_20 = df.close.rolling(window=20).mean()
rolling_mean_50 = df.close.rolling(window=50).mean()

# plot
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(df.index, df.close, label='TSLA')
plt.plot(df.index, rolling_mean_20, label='TSLA 20 Day SMA')
plt.plot(df.index, rolling_mean_50, label='TSLA 50 Day SMA')
plt.legend(loc='upper left')
plt.show()
```

## Exponential Moving Average (EWMA)

```python
# series
exp_20 = df.close.ewm(span=20, adjust=False).mean()
exp_50 = df.close.ewm(span=50, adjust=False).mean()

# plot
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(df.index, df.close, label='TSLA')
plt.plot(df.index, exp_20, label='TSLA 20 Day EMA')
plt.plot(df.index, exp_50, label='TSLA 50 Day EMA')
plt.legend(loc='upper left')
plt.show()
```

## Trend-change signals

```python
def plot_graph(stock_price_series
               , avg_type="ma"
               , period = "day"
               , st=20
               , lt=50
               , identify_signals=True
               , figsize=(20,10)
              ):
    
    """
    Y
    """
    
    assert avg_type=="ma" or avg_type=="ewma", "avg_type must be <<ma>> or <<ewma>>"

    # Data
    X = stock_price_series.index
    Y1 = stock_price_series.values
    
    if avg_type=="ma":
        Y2 = stock_price_series.rolling(window=st).mean() #stma
        Y3 = stock_price_series.rolling(window=lt).mean() #ltma
    else:
        Y2 = stock_price_series.ewm(span=st, adjust=False).mean() #stma
        Y3 = stock_price_series.ewm(span=lt, adjust=False).mean() #ltma

    # Create Figure 
    fig, ax = plt.subplots(figsize=figsize)

    # Plot series
    ax.plot(X, Y1, lw=2, label="Close Price", color='dodgerblue')
    ax.plot(X, Y2, lw=2, label="{} {} {}".format(st,period.capitalize(),avg_type.upper()), color='darkorange')
    ax.plot(X, Y3, lw=2, label="{} {} {}".format(lt,period.capitalize(),avg_type.upper()), color='darkmagenta')

    # x axis
    ax.xaxis.set_major_locator(MonthLocator())
    ax.xaxis.set_major_formatter(DateFormatter("%Y-%m"))
    ax.tick_params(which='major',width=1.0,length=10)
    fig.autofmt_xdate()

    # Titles, labels, legend and grid
    ax.set_title("technical analysis", fontsize=20, verticalalignment='bottom')
    ax.set_xlabel("date")
    ax.set_ylabel("stock price")
    ax.legend(loc='upper left')
    plt.xlim(X.min(),X.max())
    ax.grid(linestyle=":", linewidth=0.5, color='gray', zorder=-10)

    # Signals
    if identify_signals == True:

        # buy signals: short term average exceeds the long term average
        buy_signals = (Y2 > Y3) & (Y2.shift(1) <= Y3.shift(1))

        # sell signals: short term average dips the long term average
        sell_signals = (Y2 < Y3) & (Y2.shift(1) >= Y3.shift(1))

        ax = plt.scatter(x = Y2[buy_signals].index,
                         y = Y2[buy_signals]*0.5 + Y3[buy_signals]*0.5,
                         s = 1,
                         linewidth = 35,
                         edgecolor='green',
                         facecolors='none',
                         clip_on = True,
                         zorder =10)

        ax = plt.scatter(x = Y2[sell_signals].index,
                         y = Y2[sell_signals]*0.5 + Y3[sell_signals]*0.5,
                         s = 1,
                         linewidth = 35,
                         edgecolor='red',
                         facecolors='none',
                         clip_on = True,
                         zorder =10)


    return plt.show()
```

```python
df_sup = df[df.index>"2020-01-01"]

plot_graph(df_sup["close"]
           ,figsize=(20,10)
          )

plot_graph(df_sup["close"]
           ,avg_type="ewma"
           ,figsize=(20,10)
          )
```
