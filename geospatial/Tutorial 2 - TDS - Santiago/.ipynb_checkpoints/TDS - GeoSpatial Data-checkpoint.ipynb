# Plotting Santiago, Chile
## Load the data

```python
# set style
sns.set(style='whitegrid', palette='pastel', color_codes=True)
sns.mpl.rc('figure', figsize=(10,6))

# import shapes
shp_path = 'Comunas_RM_Mapas_Vectoriales/Comuna.shp'
sf = shp.Reader(shp_path)
print('Number of shapes imported:',len(sf.shapes()))

# example
print('\nExamples:')
print(sf.records()[0])
print(sf.records()[1])

# example
df = read_shapefile(sf)
print(df.shape)
df.head()
```

## Ploting a single shape

```python
# example 1
comuna = 'SANTIAGO'
com_id = df[df.NOM_COMUNA == comuna].index.tolist()[0]
plot_shape(sf, com_id, comuna)

print(type(sf))
print(type(com_id))

# example 2
comuna = 'LA FLORIDA'
com_id = df[df.NOM_COMUNA == comuna].index.tolist()[0]
plot_shape(sf, com_id, comuna)
```

## Plot map within limiting coordinates

```python
# no limiting coordinates
plot_map(sf=sf)

# with limiting coordinates
y_lim = (-33.7,-33.3) # lat
x_lim = (-71,-70.25) # long
plot_map(sf=sf,
         x_lim=x_lim,
         y_lim=y_lim,
         highlight=True)
```

## Plot map within limiting coordinates, border highlight and fill

```python
# example 1 (by code)
comuna_id = [0, 1, 2, 3, 4, 5, 6]
plot_map_v2(title="Multiple Shapes",
            comuna=comuna_id,
            sf=sf)

# example 2 (by name)
comunas = ['QUILICURA',
           'Estación Central',
           'Santiago',
           'San Bernardo',
           'Melipilla',
           'Recoleta',
           'Pedro Aguirre Cerda',
           'Lo Espejo',
           'Independencia',
           'Cerro Navia']
color = '#00FF0030'

plot_map_v2(title='Haitianos',
            comuna=comunas,
            sf=sf)
```

```python
# neighborhood names

center = ['santiago']
west = ['providencia', 'vitacura', 'lo barnechea', 'las condes', 'la reina', 'nunoa']
southeast = ['san jose de maipo', 'macul', 'la florida', 'penalolen',]
south = ['alhue', 'calera de tango', 'buin', 'isla de maipo', 'el bosque', 'paine', 'la granja', 'pedro aguirre cerda', 'lo espejo', 'puente alto', 'san joaquin', 'san miguel', 'pirque', 'san bernardo', 'san ramon', 'la cisterna', 'talagante', 'la pintana', ]
north = ['tiltil', 'renca', 'quinta normal', 'cerro navia', 'lampa', 'conchali', 'recoleta', 'independencia', 'colina', 'huechuraba', 'quilicura', ]
east = ['san pedro', 'maria pinto', 'el monte', 'curacavi', 'padre hurtado', 'melipilla', 'estacion central', 'maipu', 'lo prado', 'pudahuel', 'penaflor', 'cerrillos', ]
        
regions = [['Central','y',[conv_comuna(i) for i in center]],
           ['West','m',[conv_comuna(i) for i in west]],
           ['Southeast','c',[conv_comuna(i) for i in southeast]],
           ['South','b',[conv_comuna(i) for i in south]],
           ['North','r',[conv_comuna(i) for i in north]],
           ['East','g',[conv_comuna(i) for i in east]]]

# plot
for region in regions:
    plot_map_v2(sf=sf,title=region[0],fill_color=region[1],comuna=region[2])
```

# Plotting USA

```python
# import data
shp_path = "cb_2017_us_state_5m/cb_2017_us_state_5m.shp"
sf_us = shp.Reader(shp_path)

# read data
df_us = read_shapefile(sf_us)
```

```python
region_name = "California"
plot_shape(sf_us
           , df_us[df_us["NAME"]==region_name].index[0]
           , region_name
          )

region_name = "Texas"
plot_shape(sf_us
           , df_us[df_us["NAME"]==region_name].index[0]
           , region_name
          )

region_name = "Florida"
plot_shape(sf_us
           , df_us[df_us["NAME"]==region_name].index[0]
           , region_name
          )
```

```python
# US
y_lim = (15,80) # lat 
x_lim = (-170,-60) # long

plot_map(sf=sf_us,
         x_lim=x_lim,
         y_lim=y_lim,
         figsize=(15,10))
```

```python
# contintental US
state_id = [0, 10, 3, 5, 6, 7, 8, 30]
plot_map_v2(title='US States',
            comuna=state_id,
            sf=sf_us,
            x_lim=x_lim,
            y_lim=y_lim,
            fill_color = 'r',
            figsize = (15,9))
```
